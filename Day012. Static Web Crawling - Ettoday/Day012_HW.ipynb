### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
re = requests.get(url)

soup = BeautifulSoup(re.text, "html5lib")
for d in soup.find(class_="part_list_2").find_all("h3"):
    print(d.find(class_="date").text, d.find(class_="tag").text, d.find_all("a")[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
news = dict()
for d in soup.find(class_="part_list_2").find_all("h3"):
    group = d.find(class_="tag").text
    title = d.find(class_="date").text + " " + d.find_all("a")[-1].text
    if group not in news:
        news[group] = [title]
    else:
        news[group] += [title]


for group in news:
    print(group, news[group])
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
max_number = -1
max_group = ''
for group in news:
    number = len(news[group])
    if number > max_number:
        max_number = number
        max_group = group
        
print(max_group, max_number)
```
