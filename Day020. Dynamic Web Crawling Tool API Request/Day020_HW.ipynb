```python
count = 0
output = dict()
rows = dict()
for item in table.find_all('td'):
    if 'style' in item.attrs and item['style'] == 'white-space:nowrap;':
        if rows:
            output[key] = rows
        
        rows = dict()
        count = 0
        key = item.string
        continue
    
    if item.string == '\xa0':
        continue
    
    if len(item.attrs) == 0:
        if count % 2 == 0:
            row_key = item.string
        else:
            rows[row_key] = item.string
        
        count += 1
        
print(output)
```
